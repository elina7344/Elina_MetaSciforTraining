```python
'''
INHERITANCE:
- Used to inherit properties and behaviors of one class to another.
- The class that inherits another class is called Child Class.
- The class that gets inherited is called the Base Class or Parent Class.

TYPES OF INHERITANCES:
- Single: where a child class inherits attributes and methods from only one parent class.
- Multiple: Child can inherit methods from two or more than two parent classes.
- Mutilevel: Child inherits from one parent class and the parent inherits from another grandparent class.
- Hierachical: contains multiple derived classes from a single base class
- Hybrid: Combination of multilevel and Multiple inheritances
'''
#Base /Parent class
class Plants:
  def Photosynthesis(self):
    print("Leaves of plants that do photosynthesis are green")

class Huge:
  def heights(self):
    print("Trees have huge heights")

#Child/derived Class

#Multiple Inheritance
class Trees(Plants, Huge):
  def __init__(self,name,habitat):
    self.name=name
    self.habitat=habitat

#Single Inheritance
class GoodForSkin(Plants):
  def Neem(self):
    print("Neem is a plant which is good for skin")
  def aloevera(self):
    print("Aloevera is a plant which is good for skin")

#Multilevel Inheritance
class Herbs(GoodForSkin):
  def Termeric(self):
    print("Termeric is a herb which is good for skin")

T=GoodForSkin()
T.aloevera()
A=Herbs()
A.Termeric()
T1=Trees("Silver Oak","Uttarakhand")
T2=Trees("Maple","Canada,Kashmir")
T3=Trees("Laburnum","Hillside,Plains")

print(T1.name,"|",T2.name,"|",T3.name)
```

```python
'''
METHOD OVERRIDING:
OOPS language allows a subclass or child class to provide a specific
implementation of a method that is already provided by one of its super-classes
or parent classes.

Example: in 2011 you didn’t have a phone but your father had a nokia then in
2020 you bought a oneplus. In 2011 you used father’s nokia to communicate
now that in 2024 you have overriden his phone with oneplus you can use your
own phone.
'''
#Defining parent
class Parent():
  def __init__(self):
    self.value = "Dad has Nokia"
  def show(self):
    print(self.value)

# Defining child class
class Child(Parent):
  def __init__(self):
    super().__init__() # Call parent constructor
    self.value = "Son has One Plus"
  def show(self):
    print(self.value)

# Driver's code
obj1 = Parent()
obj2 = Child()
obj1.show() # Should print "Inside Parent"
obj2.show() # Should print "Inside Child"
```

```python
'''
ENCAPSULATION:
- Process of bundling attributes and methods within a single unit
- According to the principle of data encapsulation, the data members that describe an object are hidden from the environment external to the class.
- They can only be accessed through the methods within the same class.
- Methods themselves on the other hand are accessible from outside class context.
- Hence, object data is said to be encapsulated by the methods.
- In this way, encapsulation prevents direct access to the object data.

Public Access Modifier:
- It is the default for member variables and methods for a class.
- Here, the data member and methods are accessible outside or inside of the class.

Private Access Modifier:
- A double underscore prefix denotes protected members before their name.
- Here, the data member and methods are accessed only within the class.

Protected Access Modifier:
- A single underscore prefix denotes protected members before their name.
- Here, the data member and methods are accessed only within the class and the class that is inherited from the protected class.
'''
#Public Access Modifiers
class MyClass:
  def __init__(self, value):
    self.value = value  # Public attribute
  def show_value(self):  # Public method
    print(self.value)
# Usage
obj = MyClass(10)
obj.show_value()  # Accessing public method
print(obj.value)  # Accessing public attribute


#Private Access Modifiers
class MyClass2:
    def __init__(self, value):
        self.__value = value  # Private attribute
    def __show_value(self):  # Private method
        print(self.__value)
    def public_method(self):  # Public method
        self.__show_value()  # Accessing private method within the class

# Usage
obj = MyClass2(20)
# obj.__show_value()  # This would raise an AttributeError
# print(obj.__value)  # This would raise an AttributeError
obj.public_method()  # Accesses private method through a public method


#Protected Access Modifier
class BaseClass:
    def __init__(self, value):
        self._value = value  # Protected attribute
    def _show_value(self):  # Protected method
        print(self._value)

class SubClass(BaseClass):
    def display(self):
        self._show_value()  # Accessing protected method from the subclass

# Usage
obj = SubClass(30)
obj.display()  # Accessing protected method via public method in subclass
print(obj._value)  # Accessing protected attribute directly (not recommended)
```
